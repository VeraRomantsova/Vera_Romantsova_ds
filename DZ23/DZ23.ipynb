# Как критики относятся к спортивным играм?

1) Нужно сформулировать нулевую и альтернативную гипотезы;
H0: средняя оценка критиками спортивных игр не отличается от оценок других жанров
H1: средняя оценка критиками спортивных игр различается от оценок других жанров
2) Определить пороговый уровень статистической значимости;
возьмем стандартный уровень в 0.05
3) Сколько у нас выборок (какой вариант теста используем)?
используем двухвыборочный тест
3) Нужен ли односторонний, либо двухсторонний вариант теста?
используем двухсторонний вариант теста (и негативная и позитивная оценка)
5) Определить, равна ли дисперсия групп.
используем критерий Уэлча

```python
# Создадим датасет только с оценками по спортивным играм
data_sport = data1[['Critic_Score']].loc[data1.Genre == 'Sports'].dropna()
# Создадим датасет только с оценками по играм остальных жанров
data_no_sport = data1[['Critic_Score']].loc[data1.Genre != 'Sports'].dropna()
```

Примечание: решила удалить строки без оценки и не заменять их на средние, например по жанру, так как на мой взгляд, это негативно повлияет на результаты сравнения. 

```python
#Задаем уровень значимости
alpha = 0.05
#Применяем двухсторонний двухвыборочный критерий Стьюдента
result = stats.ttest_ind(data_sport['Critic_Score'], data_no_sport['Critic_Score'], equal_var=False)
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, средняя оценка критиками спортивных игр различается от оценок других жанров')
else:
    print('Не отвергаем нулевую гипотезу')
```

Вывод: p_value > 0.05 -> не отвергаем нулевую гипотезу о равенстве средних оценок спортивных и остальных игр, 
    при этом средняя оценка по спортивным играм немного выше средней оценки игр других жанров.

# 2. Критикам нравятся больше игры на PC или на PS4?
1) Нужно сформулировать нулевую и альтернативную гипотезы;
H0: средняя оценка критиками игр на PC и PS4 одинакова
H1: средняя оценка критиками игр на PC и PS4 различается
2) Определить пороговый уровень статистической значимости;
возьмем стандартный уровень в 0.05
3) Сколько у нас выборок (какой вариант теста используем)?
используем двухвыборочный тест
3) Нужен ли односторонний, либо двухсторонний вариант теста?
используем двухсторонний вариант теста (так как нас интересует игры на какой платформе больше нравится критикам)
5) Определить, равна ли дисперсия групп.
используем критерий Уэлча

```python
# Создадим датасет только с оценками критиков по играм на PC
data_PC = data1[['Critic_Score']].loc[data1.Platform == 'PC'].dropna()
# Создадим датасет только с оценками критиков по играм на PS4
data_PS4 = data1[['Critic_Score']].loc[data1.Platform == 'PS4'].dropna()
```

Примечание: решила удалить строки без оценки и не заменять их на средние, например по жанру, так как на мой взгляд, это негативно повлияет на результаты сравнения.

```python
#Задаем уровень значимости
alpha = 0.05
#Применяем односторонний двухвыборочный критерий Стьюдента
result = stats.ttest_ind(data_PC['Critic_Score'], data_PS4['Critic_Score'], equal_var=False)
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, средняя оценка критиками игр на PC различается от оценок игр на PS4')
else:
    print('Не отвергаем нулевую гипотезу')
```

Вывод: p_value < 0.05 -> отвергаем нулевую гипотезу о равенстве средних оценок критиков игр на PC и PS4, при этом средняя оценка на PC ниже средней оценки игр на PS4. С высокой долей статзначимости можем утверждать, что критикам больше нравятся игры на PS4.

# 3) Критикам больше нравятся стрелялки или стратегии?
1) Нужно сформулировать нулевую и альтернативную гипотезы; H0: средняя оценка критиками "стрелялок" и "стратегий" одинакова H1: средняя оценка критиками "стрелялок" и "стратегий" различается 2) Определить пороговый уровень статистической значимости; возьмем стандартный уровень в 0.05 3) Сколько у нас выборок (какой вариант теста используем)? используем двухвыборочный тест 3) Нужен ли односторонний, либо двухсторонний вариант теста? используем двухсторонний вариант теста (так как нас интересует какой вид игры больше нравится критикам) 5) Определить, равна ли дисперсия групп. используем критерий Уэлча

```python
# Создадим датасет только с оценками критиков по играм на PC
data_Shooter = data1[['Critic_Score']].loc[data1.Genre == 'Shooter'].dropna()
# Создадим датасет только с оценками критиков по играм на PS4
data_Strategy = data1[['Critic_Score']].loc[data1.Genre == 'Strategy'].dropna()
```

Примечание: решила удалить строки без оценки и не заменять их на средние, например по жанру, так как на мой взгляд, это негативно повлияет на результаты сравнения.

```python
#Задаем уровень значимости
alpha = 0.05
#Применяем односторонний двухвыборочный критерий Стьюдента
result = stats.ttest_ind(data_Shooter['Critic_Score'], data_Strategy['Critic_Score'], equal_var=False)
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, средняя оценка критиками "стрелялок" различается от оценок "стратегий"')
else:
    print('Не отвергаем нулевую гипотезу')
```

Вывод: p_value > 0.05 -> не отвергаем нулевую гипотезу о равенстве средних оценок критиков игр на PC и PS4.
С недостаточным уровнем статзначимости можем сказать, что средняя оценка стратегий выше средней оценки стрелялок.

```python
data = pd.read_csv('spam.txt')
```

```python
data.head()
```

```python
import re
import nltk
nltk.download('stopwords')
nltk.download('wordnet')

from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))

from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
def normalize_word(string_):
    string_ = string_.lower() #Приводим к нижнему регистру
    string_ = re.sub('[\W_]+',' ', string_) # уберем все служебные символы
    string_ = string_.split() # разделим все на слова
    string_ = [word for word in string_ if word not in stopwords_set] #убираем стоп-слова
    string_ = [wordnet_lemmatizer.lemmatize(word) for word in string_] #приведем слова к леммам
    string_ = ' '.join(map(str, string_)) #преобразуем в строку
    return string_
data['Message'] = data.Message.apply(normalize_word)
```

```python
data.head()
```

```python
# Преобразуем все сообщения в вектора TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(data.Message)
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=None)
tfidf_matrix.head()
```

```python
# Разделим данные на тестовые и тренировочные 30\70 
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, data['Category'], test_size=0.30, random_state=42)
```

```python
# Построим модель логистической регресии, установим random_state=42
lr = LogisticRegression(random_state=42)
```

```python
lr.fit(X_train, y_train) #Обучаем модель на тренировочных данных
```

```python
lr.predict(X_test) #Строим прогноз на тестовых данных
```

```python
# Оценим ее точность на тестовых данных
from sklearn.metrics import accuracy_score
accuracy_score(y_test, lr.predict(X_test))
```

Вы видим, что доля правильных ответов чуть более 0.95

```python
# Опишем результаты при помощи confusion_matrix
from sklearn.metrics import confusion_matrix
cnf_matrix = confusion_matrix(y_test, lr.predict(X_test))
cnf_matrix
```
