Задание 3 - предлагаемая методика работы с пропусками
Столбцы с пропусками не стоит удалять из DF, так как присутствие данных по всех столбцах достаточно (пропусков не более 25%).

Я подозреваю, что наличие пропусков в первую очередь связано с двумя причинами:
    1) лошадь поступила в тяжелом состоянии - все обследование сделать не получилось или лошадь умерла и обследование закончить не успели
    2) при обследовании получили информацию, что позволило поставить диагноз и остальные исследования не имели смысла.
Я думаю, что эти пропуски относятся к MAR, поэтому замена пропусков не должна привести к существенному искажению результатов.

Попробую заменить пропуски и уже потом при наличии строк с пустыми значениями удалю их (при недостаточности данных в этой строке для относительно корректной замены значений)

```python
#Заполним пропуски по ректальной температуре медианой (только для случаев с повышенной температурой конечностей) - "Имитация жара"
high_temp_ex = cleaned_df['temperature_of_extremities'] == 2.0
median_high_temp_ex = cleaned_df[high_temp_ex]['rectal_temperature'].median()
cleaned_df.loc[high_temp_ex & cleaned_df.rectal_temperature.isna(),'rectal_temperature'] = median_high_temp_ex
```

```python
# Заполним пропуски по ректальной температуре медианой по категориям температуры конечностей и времени наполнения капилляров:
cleaned_df.rectal_temperature.fillna(cleaned_df.groupby(['temperature_of_extremities', 'capillary_refill_time']).rectal_temperature.transform('median'), inplace=True)
```

```python
# Заполним пропуски по ректальной температуре медианой по категории времени наполнения капилляров или температуры конечностей:
cleaned_df.rectal_temperature.fillna(cleaned_df.groupby(['capillary_refill_time']).rectal_temperature.transform('median'), inplace=True)
cleaned_df.rectal_temperature.fillna(cleaned_df.groupby(['temperature_of_extremities']).rectal_temperature.transform('median'), inplace=True)
```

```python
# Заполним пропуски по пульсу медианой для темно-синюшных слизистых, пониженного периферического пульса, холодных конечностей "Имитация шока":
low_temp_ex = cleaned_df['temperature_of_extremities'] == (4.0 or 3.0)
low_pp = cleaned_df['peripheral_pulse'] == (3.0 or 4.0)
dark_cyan = cleaned_df['mucous_membranes'] == 6.0

cleaned_df.loc[low_temp_ex & low_pp & dark_cyan & cleaned_df.pulse.isna(),'pulse'] = cleaned_df[low_temp_ex & low_pp & dark_cyan]['pulse'].median()
cleaned_df.loc[low_temp_ex & low_pp & cleaned_df.pulse.isna(),'pulse'] = cleaned_df[low_temp_ex & low_pp]['pulse'].median()
cleaned_df.loc[low_pp & dark_cyan & cleaned_df.pulse.isna(),'pulse'] = cleaned_df[low_pp & dark_cyan]['pulse'].median()
cleaned_df.loc[low_temp_ex & dark_cyan & cleaned_df.pulse.isna(),'pulse'] = cleaned_df[low_temp_ex & dark_cyan]['pulse'].median()
```

```python
#Заполним пропуски по пульсу (только для случаев с болью) - "Имитация боли"
high_pain = cleaned_df['pain'] == (4.0 or 5.0)
high_median = cleaned_df[(cleaned_df['pain'] == (4.0 or 5.0))]['pulse'].median()
cleaned_df.loc[high_pain & cleaned_df.pulse.isna(),'pulse'] = high_median
```

```python
# Заполним пропуски по пульсу медианой по категориям (температура конечностей):
cleaned_df.pulse.fillna(cleaned_df.groupby(['temperature_of_extremities']).pulse.transform('median'), inplace=True)
```

```python
high_temp = cleaned_df['rectal_temperature'] > 37.8
low_temp = cleaned_df['rectal_temperature'] <= 37.8
mode_high_temp = cleaned_df[high_temp]['temperature_of_extremities'].mode()[0]
mode_low_temp = cleaned_df[low_temp]['temperature_of_extremities'].mode()[0]

#Заполним пропуски по температуре конечностей модой (только для случаев с повышенной ректальной температурой) - "Имитация жара"
cleaned_df.loc[high_temp & cleaned_df.temperature_of_extremities.isna(),'temperature_of_extremities'] = mode_high_temp

#Заполним пропуски по температуре конечностей модой (только для случаев с пониженной ректальной температурой) - "Имитация шока"
cleaned_df.loc[low_temp & cleaned_df.temperature_of_extremities.isna(),'temperature_of_extremities'] = mode_low_temp
```

```python
mode_cap_refill_1 = cleaned_df[high_temp]['capillary_refill_time'].mode()[0]
mode_cap_refill_2 = cleaned_df[low_temp]['capillary_refill_time'].mode()[0]
# Заполним пропуски по времени заполнения капилляров модой (только для случаев с повышенной ректальной температурой)
cleaned_df.loc[high_temp & cleaned_df.capillary_refill_time.isna(),'capillary_refill_time'] = mode_cap_refill_1
# Заполним пропуски по по времени заполнения капилляров модой (только для случаев с пониженной ректальной температурой)
cleaned_df.loc[low_temp & cleaned_df.capillary_refill_time.isna(),'capillary_refill_time'] = mode_cap_refill_2
```

```python
# Заполним пропуски по периф. пульсу (корреляция с временем заполнения капилляров, слиз. оболочками):
# Корреляцию с пульсом не стала смотреть ввиду высокой дисперсии пульса.
cleaned_df['peripheral_pulse'] = cleaned_df.groupby(['capillary_refill_time','mucous_membranes'])['peripheral_pulse'].apply(lambda x: x.fillna(x.mode()[0]))
cleaned_df['peripheral_pulse'] = cleaned_df.groupby(['capillary_refill_time'])['peripheral_pulse'].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
# Заполним пропуски по слизистым оболочкам (корреляция с временем заполнения капилляров, периф пульсом):
# Корреляцию с пульсом не стала смотреть ввиду высокой дисперсии пульса.
cleaned_df['mucous_membranes'] = cleaned_df.groupby(['capillary_refill_time','peripheral_pulse'])['mucous_membranes'].apply(lambda x: x.fillna(x.mode()[0]))
cleaned_df['mucous_membranes'] = cleaned_df.groupby(['capillary_refill_time'])['mucous_membranes'].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
#Заполним пропуски по вздутию (только для случаев с высокой болью) - "Имитация боли"
high_pain4 = cleaned_df['pain'] == 4.0
high_pain5 = cleaned_df['pain'] == 5.0
high_pain_mode4 = cleaned_df[high_pain4]['abdominal_distension'].mode()[0]
high_pain_mode5 = cleaned_df[high_pain5]['abdominal_distension'].mode()[0]
cleaned_df.loc[high_pain4 & cleaned_df.abdominal_distension.isna(),'abdominal_distension'] = high_pain_mode4
cleaned_df.loc[high_pain5 & cleaned_df.abdominal_distension.isna(),'abdominal_distension'] = high_pain_mode5
```

```python
mode_high_temp_pain = cleaned_df[high_temp]['pain'].mode()[0]
mode_low_temp_pain = cleaned_df[low_temp]['pain'].mode()[0]

#Заполним пропуски по температуре конечностей модой (только для случаев с повышенной ректальной температурой) - "Имитация жара"
cleaned_df.loc[high_temp & cleaned_df.pain.isna(),'pain'] = mode_high_temp_pain

#Заполним пропуски по температуре конечностей модой (только для случаев с пониженной ректальной температурой) - "Имитация шока"
cleaned_df.loc[low_temp & cleaned_df.pain.isna(),'pain'] = mode_low_temp_pain
```

```python
#Заполним пропуски по вздутию модой для категории боли, периф пульса
cleaned_df['abdominal_distension'] = cleaned_df.groupby(['pain'])['abdominal_distension'].apply(lambda x: x.fillna(x.mode()[0]))
```
