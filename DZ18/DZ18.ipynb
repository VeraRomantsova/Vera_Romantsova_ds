Смоделировать игру против лотерейного автомата типа “777”. 
Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. 
При некоторых значениях числа игрок получает выигрыш (см. справа)

Выгодна ли игра игроку?

Сколько в среднем приобретает
или теряет игрок за одну игру?

777: 200 руб.
999: 100 руб.
555: 50 руб.
333: 15 руб.
111: 10 руб.
*77: 5 руб.
**7: 3 руб.
*00: 2 руб.
**0: 1 руб.

– любая цифра
Дополнительное задание повышенной сложности. 
Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

```python
# Создаем список "удачных" номеров - сначала записываю редко выпадаемые номера
lucky_all = {'777':200,
             '999':100,
             '555':50,
             '333':15,
             '111':10,
             '077':5,
             '177':5,
             '277':5,
             '377':5,
             '477':5,
             '577':5,
             '677':5,
             '877':5,
             '977':5,
             '000':2,
            '100':2,
            '200':2,
            '300':2,
            '400':2,
            '500':2,
            '600':2,
            '700':2,
            '800':2,
            '900':2}
lucky_all
```

```python
# Генерирую список номеров, заканчивающихся на 7
import re
import numpy as np
lucky_3 = []
numbers = 1000
for i in range(numbers):
    if i < 10 and re.findall(r'\d{2}7',('00'+str(i))) and ('00'+str(i)) not in lucky_3:
        lucky_3.append(('00'+str(i)))
    elif i < 100 and re.findall(r'\d{2}7',('0'+str(i))) and ('0'+str(i)) not in lucky_3:
            lucky_3.append(('0'+str(i)))
    elif re.findall(r'\d{2}7',str(i)) and str(i) not in lucky_3:
                lucky_3.append(str(i))
lucky_3
```

```python
#Записываю номера, заканчивающиеся на 7 (сформированные в прошлом шаге) в список всех удачных номеров и назначаю выигрыш в 3
for x in lucky_3:
    if x not in lucky_all.keys():
        lucky_all[x] = 3
lucky_all
```

```python
# Генерирую список номеров, заканчивающихся на 0
lucky_0 = []
numbers = 1000
for i in range(numbers):
    if i < 100 and re.findall(r'\d{2}0',('0'+str(i))) and ('0'+str(i)) not in lucky_0:
            lucky_0.append(('0'+str(i)))
    elif re.findall(r'\d{2}0',str(i)) and str(i) not in lucky_0:
                lucky_0.append(str(i))
lucky_0
```

```python
#Записываю номера, заканчивающиеся на 0 (сфорсированные в прошлом шаге) в список всех удачных номеров и назначаю выигрыш в 1
for x in lucky_0:
    if x not in lucky_all.keys():
        lucky_all[x] = 1
lucky_all
```

```python
# np.random.seed(None)
trials = 10000000    # Количество попыток (они же траты на попытку, т.к. стоимость попытки = 1)
money = 0    # Сумма выигрышей
successes = 0 # Количество выигрышей

for i in range(trials):
    random_num = np.random.randint(0, 1000)    # Случайное число от 0 до 999
    if random_num <10 and ('00'+str(random_num)) in lucky_all.keys(): #Если <10 - добавляю 00
        money += int(lucky_all[('00'+str(random_num))])
        successes +=1
    elif random_num <100 and ('0'+str(random_num)) in lucky_all.keys():
        money += int(lucky_all[('0'+str(random_num))])
        successes +=1
    elif str(random_num) in lucky_all.keys():
        money += int(lucky_all[str(random_num)])
        successes +=1
p_monte_carlo = successes / trials    # Вероятность выигрыша одной попытки из моделирования (я бы скорее назвала положительного исхода)
print(successes, trials, p_monte_carlo, money)
```

```python
# Выгода на одну попытку по Монте-Карло (моделирование, (сумма выигрышей - сумма затрат на попытки) / кол-во попыток)
s = (money - trials)/trials
s
```

```python
# Мат ожидание значения выигрыша (одной попытки)
p = (1/1000)*(200) + (1/1000)*(100) + (1/1000)*(50) + (1/1000)*(15) + (10)*(1/1000) + (5)*(9/1000) + (2)*(10/1000) + (3)*(90/1000) + (1)*(90/1000)
# Тогда ожидаемая теоретическая выгода (выигрыш p минус стоимость одной попытки) составит:
m = p - 1
m
```
