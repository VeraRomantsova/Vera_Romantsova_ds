### Задание 7

```python
# Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом
```

```python
# 1) Нужно сформулировать нулевую и альтернативную гипотезы; 
# H0: среднее время, проведенное сотрудниками с высокой з.пл не отличается от среднего времени сотрудников с низким окладом
# H1: среднее время, проведенное сотрудниками с высокой з.пл отличается от среднего времени сотрудников с низким окладом
# 2) Определить пороговый уровень статистической значимости; возьмем стандартный уровень в 0.05 
# 3) Сколько у нас выборок (какой вариант теста используем)? используем двухвыборочный тест 
# 4) Нужен ли односторонний, либо двухсторонний вариант теста? используем двухсторонний вариант теста (так как интересует отклонение в ту и в другую стороны)
# 5) Определить, равна ли дисперсия групп. используем критерий Уэлча, распределения неоднородные
```

```python
data_low = data[['average_montly_hours']].loc[data.salary == 'low']
data_high = data[['average_montly_hours']].loc[data.salary == 'high'] 
```

```python
data_low.average_montly_hours.plot(kind='hist', 
                     bins=11,
                     title='average_montly_hours by workers with low salary',
                     xlabel='Salary',
                     ylabel='Count')
```

```python
data_high.average_montly_hours.plot(kind='hist', 
                     bins=11,
                     title='average_montly_hours by workers with high salary',
                     xlabel='Salary',
                     ylabel='Count')
```

```python
import scipy.stats as stats
#Задаем уровень значимости
alpha = 0.05
#Применяем двухсторонний двухвыборочный критерий Стьюдента
result = stats.ttest_ind(data_low['average_montly_hours'], data_high['average_montly_hours'], equal_var=False)
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, среднее время, проведенное сотрудниками с высокой з.пл отличается от среднего времени сотрудников с низким окладом')
else:
    print('Не отвергаем нулевую гипотезу')
```

Вывод: с высокой долей стат значимости не отвергаем нулевую гипотезу о равенстве времени сотрудников с разным уровнем з.пл.
Из данных t-статистики (>0) можем судить о том, что сотрудники с низким окладом проводят чуть больше времени на работе.

### Задание 8

```python
#Рассчитать следующие показатели среди уволившихся и неуволившихся сотрудников (по отдельности):
data_active = data[['satisfaction_level','left','promotion_last_5years','number_project']].loc[data.left == 1]
data_dismissal = data[['satisfaction_level','left','promotion_last_5years','number_project']].loc[data.left == 0]
```

```python
#Доля сотрудников с повышением за последние 5 лет
def promotion(data_):
    """Находим долю сотрудников с повышением за 5 лет в выборке"""
    return round(data_[['promotion_last_5years']].loc[data_.promotion_last_5years == 1].count() / data_[['promotion_last_5years']].count(),3)
print(promotion(data_active)) #Доля работающих сотрудников с повышением за последние 5 лет
print(promotion(data_dismissal)) #Доля сотрудников с повышением за последние 5 лет среди уволенных 
```

```python
#Средняя степень удовлетворенности
def satisfaction_level(data_):
    """Находим cреднюю степень удовлетворенности в выборке"""
    return data_.satisfaction_level.mean()
print(round(satisfaction_level(data_active),3)) #Средняя степень удовлетворенности среди работающих сотрудников
print(round(satisfaction_level(data_dismissal),3)) #Средняя степень удовлетворенности среди уволенных сотрудников
```

```python
#Среднее количество проектов
def number_project(data_):
    """Находим cреднее количество проектов в выборке"""
    return data_.number_project.mean()
print(round(number_project(data_active),3)) #Среднее количество проектов среди работающих сотрудников
print(round(number_project(data_dismissal),3)) #Среднее количество проектов среди уволенных сотрудников
```

```python
# разделяем выборку на тренировочную и тестовую
data_class = data.left
data_lda = data[['satisfaction_level','last_evaluation','number_project','average_montly_hours','time_spend_company','Work_accident','promotion_last_5years']]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_lda, data_class, test_size=0.2)
```

```python
# импортируем LDA-функционал
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
#создаем пустую LDA
lda = LinearDiscriminantAnalysis()
```

```python
#обучаем данные
lda.fit(X_train, y_train)
```

```python
# делаем прогноз на тестовой выборке
lda.predict(X_test)
```

```python
#Оценим точность прогноза
from sklearn.metrics import accuracy_score
accuracy_score(y_test, lda.predict(X_test))
```
