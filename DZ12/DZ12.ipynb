Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
  - оценка 2 и меньше - низкий рейтинг
  - оценка 4 и меньше - средний рейтинг
  - оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
data = pd.read_csv('ratings.csv')
data.head() #Читаем первые пять строк датафрейма
```

```python
def class_(rating):
    """ Присвает класс:
    низкий рейтинг - если значение rating <= 2
    средний рейтинг - если значение rating > 2 и rating <= 4
    высокий рейтинг - если значение rating >= 2
    """
    if rating <= 2:
        return "низкий рейтинг"
    if rating <= 4:
        return "средний рейтинг"
    else:
        return "высокий рейтинг"
```

```python
data['class'] = data['rating'].apply(class_) #Применим ф-ю class_ к нашему датафрейму (нет смысла применять к строке в целиком - class_ определяется только на основе даныных рейтинга)
data.head()
```

Задание 2 
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

Результат классификации запишите в отдельный столбец region.


```python
dt = pd.read_csv('keywords.csv')
dt.head(10) #Читаем первые 10 строк датафрейма
```

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
geo_data
```

```python
def regions(keyword):
    """ Для каждого города в списке городов ищем сопоставление с каждым словом в поисковом запросе keyword
    - если оно есть - возвращаем название региона этого города
    - если сопоставление не найдено ни для одного слова в поисковом запросе - возвращаем undefined"""
    for key, values in geo_data.items():
        for value in values:
            for word in keyword.split(' '):
                if value == word:
                    return(key)
    return 'undefined'
```

```python
dt['region'] = dt['keyword'].apply(regions) #Для DF с поисковыми запросами - добавляем столбец с регионом
dt.head(10)
```

```python
dt['region'].value_counts() # Вывожу группировку по кол-ву поисковых значений, содержащих значение региона
```

Задание 3 (бонусное)
Есть мнение, что "раньше снимали настоящее кино, не то что сейчас". Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл ratings.csv из базы https://grouplens.org/datasets/movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010.

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
  - для каждой строки пройдите по всем годам списка years
  - если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
  - если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец 'year'

4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца 'year' и отсортируйте результат по убыванию рейтинга

```python
years = list(range(1950,2011)) #Создаю список лет
```

```python
data = pd.read_csv('ratings.csv')

df = pd.read_csv('movies.csv')
df.head() #Читаем первые 5 строк датафрейма
```

```python
import re

def production_year(text):
    """ Пишем регулярку на поиск года в названии фильма
    если год находится и он есть в списке years - возвращаем год 
    если оба условия не срабатывают (года в списке нет или года нет совсем ) - возвращает 1900"""
    year_ = re.findall(r'\((\d{4})\)', text)
    if year_ != [] and int(year_[0]) in years:
        return int(year_[0])
    else:
        return 1900
```

```python
df['year'] = df['title'].apply(production_year) #Применяем production_year к названиям фильмов в нашем DF - и создаем столбец year
df.head()
```

```python
data_mean = data.merge(df, how='left').groupby(['year']).mean().sort_values('rating', ascending=False)
data_mean.head() #Нахожу средний рейтинг через mean() для группировки по значениям столбца 'year' 
# (для этого к rating сначала нужно приджойнить year из movies) 
# результат сортирую по убыванию среднего значения рейтинга для каждого года
```
