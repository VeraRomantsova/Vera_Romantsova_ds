```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
# Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру
def get_doc_name(number):
    """
    Находим владельца по номеру документа
    """
    for document in documents:
        if document['number'] == number:
            return document['name']

def get_doc_number(number):
    """
    Находим документ
    """
    for document in documents:
        if document['number'] == number:
            return document['number']
            
def search_document():
    """
    Оберточная функция вывода результатов по поиску владельца по номеру документа
    """
    user_input = input('Введите номер документа:')
    if user_input == get_doc_number(user_input):
        print('Владелец документа: ',get_doc_name(user_input))
    else:
        print('Документ не найден в базе')
```

```python
#Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится

def get_doc_directory(number):
    """
    Находим непустую полку по номеру документа
    """
    for key, value in directories.items():
        if number in value:
            return key
            
def search_directory():
    """
    Оберточная функция вывода результатов по поиску полки по номеру документа
    """
    user_input = input('Введите номер документа:')
    if user_input == get_doc_number(user_input):
        print('Документ хранится на полке: ',get_doc_directory(user_input))
    else:
        print('Документ не найден в базе')
```

```python
# Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам
def get_doc_type(number):
    """
    Находим тип документа по его номеру
    """
    for document in documents:
        if document['number'] == number:
            return document['type']

def search_doc_info():
    """
    Оберточная функция вывода результатов по поиску всей информации о документе по его номеру 
    """
    user_input = input('Введите номер документа:')
    if user_input == get_doc_number(user_input):
        print(f'N: {get_doc_number(user_input)}, тип: {get_doc_type(user_input)}, владелец: {get_doc_name(user_input)}, полка хранения: {get_doc_directory(user_input)}')
    else:
        print('Документ не найден в базе')
```

```python
 # Пункт 4. Пользователь по команде "ads" может добавить новую полку
def get_exist_directory(number):
    """
    Находим есть ли такая полка 
    """
    for key in list(directories.keys()):
        if number in key:
            return key

def add_directory():
    """
    Оберточная функция добавления полки и вывода результатов по текущему перечню полок 
    """
    user_input = input('Введите номер полки:')
    if user_input == get_exist_directory(user_input):
        print(f"Такая полка уже существует. Текущий перечень полок:  {','.join([key for key in directories.keys()])}")
    else:
        directories[user_input] = []
        print(f"Полка добавлена. Текущий перечень полок:  {','.join([key for key in directories.keys()])}")
```

```python
# Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)

def del_directory():
    """
    Оберточная функция удаления полки и вывода результатов по текущему перечню полок
    """
    user_input = input('Введите номер полки:')
    if user_input != get_exist_directory(user_input):
        print(f"Такой полки не существует. Текущий перечень полок:  {','.join([key for key in directories.keys()])}")
    elif directories.get(user_input, None) == []:
        del(directories[user_input])
        print(f"Полка удалена. Текущий перечень полок:  {','.join([key for key in directories.keys()])}")
    else:
        print(f"На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:  {','.join([key for key in directories.keys()])}")      
```

```python
# Пункт 1 (дополнительно). Пользователь по команде "ad" может добавить новый документ в данные
def get_document_directory(number):
    """
    Находим есть ли полка в базе
    """
    for key in directories.keys():
        if number in key:
            return key 
        
def get_info_all_docs():
    """
    Выводим все данные по всем документам
    """
    x = []
    for document in documents:
        number = document['number']
        x.append(f'N: {get_doc_number(number)}, тип: {get_doc_type(number)}, владелец: {get_doc_name(number)}, полка хранения: {get_doc_directory(number)}')
    return(x)


def add_document():
    """
    Оберточная функция добавления документа
    """
    user_input = input('Введите номер документа:')
    user_input_type = input('Введите тип документа:')
    user_input_name = input('Введите владельца документа:')
    user_input_directory = input('Введите полку для хранения:')
    if user_input_directory == get_document_directory(user_input_directory):
        documents.append({'type': user_input_type, 'number': user_input, 'name': user_input_name})
        for key, value in directories.items():
            if key == user_input_directory:
                value.append(user_input)
        print(f"Документ добавлен. Текущий перечень документов:")
        [print(x) for x in get_info_all_docs()]
    else: 
        print(f"Такой полки не существует. Добавьте полку командой as. Текущий перечень документов:")  
        [print(x) for x in get_info_all_docs()]
add_document()
```

```python
#Пункт 2 (дополнительно). Пользователь по команде "d" может удалить документ из данных

def del_doc_in_docs(user_input):
    """
    Удаление документа из списка документов
    """
    for e in range(len(documents)-1, -1, -1):
        if documents[e]['number'] == user_input:
            documents.pop(e)
            
def del_doc_in_directory(user_input):
    """
    Удаление документа из полки
    """
    for key, value in directories.items():
        if user_input in value:
            value.remove(user_input)
            
def del_document():
    """
    Оберточная функция удаления документа
    """
    user_input = input('Введите номер документа:')
    if user_input == get_doc_number(user_input):
        del_doc_in_docs(user_input)
        del_doc_in_directory(user_input)
        print(f"Документ удален. Текущий перечень документов:")
        [print(x) for x in get_info_all_docs()]
    else: 
        print(f"Документ не найден в базе. Текущий перечень документов:")  
        [print(x) for x in get_info_all_docs()]
```

```python
#Пункт 3 (дополнительно). Пользователь по команде "m" может переместить документ с полки на полку

def replace_document():
    """
    Оберточная функция перемещения документа
    """
    user_input = input('Введите номер документа:')
    user_input_directory = input('Введите полку для хранения:')
    if user_input == get_doc_number(user_input) and user_input_directory == get_exist_directory(user_input_directory):
        del_doc_in_directory(user_input)
        for key, value in directories.items():
            if key == user_input_directory:
                value.append(user_input)
        print(f"Документ перемещен. Текущий перечень документов:")
        [print(x) for x in get_info_all_docs()]
    elif user_input == get_doc_number(user_input):
        print(f"Такой полки не существует. Текущий перечень полок: {','.join([key for key in directories.keys()])}")
    else:
        print(f"Документ не найден в базе.  Текущий перечень документов:")
        [print(x) for x in get_info_all_docs()]
replace_document()
```

```python
def main():
    """
    переходим к поиску или действию согласно:
     - p" может узнать владельца документа по его номеру
     - "s" может по номеру документа узнать на какой полке он хранится
     - "l" может увидеть полную информацию по всем документам
     - "ads" может добавить новую полку
     - "ds" может удалить существующую полку из данных (только если она пустая)
    """
while True:
    user_input_ = input('Введите команду')
    if user_input_ == 'p':
        search_document()
    elif user_input_ == 's':
        search_directory()
    elif user_input_ == 'l':
        search_doc_info()
    elif user_input_ == 'ads':
        add_directory()
    elif user_input_ == 'ds':
        del_directory() 
    elif user_input_ == 'ad':
        add_document()
    elif user_input_ == 'd':
        del_document()
    elif user_input_ == 'm':
        replace_document()    
    elif user_input_ == 'q':
        break
    else:
        print('Введена неверная команда')
```
